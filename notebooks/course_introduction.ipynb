## What Is This Template?

This template combines:
- **JupyterBook**: For structuring markdown and notebooks into a book.
- **JupyterLite**: For running Python code in the browser without installation.
- **GitHub Actions**: For auto-deploying your book to a live website.
- **Badges**: For launching notebooks in JupyterLite or Google Colab.
