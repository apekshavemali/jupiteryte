# Welcome to Your JupyterBook + JupyterLite Template

This template makes it easy to create an interactive book using **JupyterBook** and **JupyterLite**. You can run Python notebooks in the browser without any setup, and your book will auto-deploy to GitHub Pages.

## Why Use This Template?

This template simplifies creating a web-based book with interactive notebooks. Key features include:
- **JupyterBook**: Organizes content into a professional book.
- **JupyterLite**: Runs Python code in the browser using Pyodide.
- **GitHub Actions**: Automatically builds and deploys your book.
- **Google Colab**: Run notebooks in the cloud with one click.

## How to Get Started

Follow these steps to set up your book:
1. Create a repository using this template on GitHub.
2. Update configuration files (`_config.yml`, `_toc.yml`).
3. Generate notebooks with `auto_notebook_creation_using_toc.py`.
4. Add your content to notebooks in `content/files/`.
5. Push changes to auto-deploy your book.

For detailed instructions, see `how_to_use.ipynb`.
