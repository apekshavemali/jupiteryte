## Prerequisites

- **GitHub Account**: To create and manage your repository.
- **Basic Git**: Know how to clone, commit, and push changes.
- **Markdown**: Basic syntax for editing content.
- **Python** (optional): For adding code to notebooks or running scripts locally.



## Tools You Don’t Need to Install

- **JupyterLite**: Runs notebooks in the browser—no installation required.
- **Google Colab**: Optional for cloud-based execution.
- **Local Setup** (optional, for testing):
  ```bash
  pip install -r requirements.txt
  chmod +x build_jupyterlite.sh
  ./build_jupyterlite.sh
  ```



## Roadmap to Create Your Book

1. **Create Repository**: Use this template on GitHub.
2. **Configure**: Update `_config.yml` and `_toc.yml` with your book’s details.
3. **Generate Notebooks**: Run `auto_notebook_creation_using_toc.py`.
4. **Add Content**: Edit notebooks in `content/files/`.
5. **Deploy**: Push to `main` for automatic GitHub Pages deployment.

### How to Apply the Content
1. **Update `_toc.yml`**:
   - Ensure `_toc.yml` matches the similar structure and change names according to your need:
     ```yaml
     format: jb-book
     root: intro
     chapters:
       - file: content/files/course_introduction
         title: Course Introduction
         sections:
           - file: content/files/course_goals
             title: Course Goals & Business Outcomes
           - file: content/files/how_to_use
             title: How to Use This Book
           - file: content/files/roadmap_prereqs
             title: Roadmap & Prerequisites
     ```
   - Save in the repository root.

2. **Generate Notebooks**:
   - Run the script to create empty notebooks:
     ```bash
     python notebooks/auto_notebook_creation_using_toc.py
     ```
   - Verify notebooks in `notebooks`:
     ```bash
     ls notebooks
     ```
