# How to Use This Template



This guide shows you how to create your own interactive book using the **JupyterBook + JupyterLite** template in the easiest way possible.

[![Launch in JupyterLite](https://jupyterlite.readthedocs.io/en/latest/_static/badge.svg)](https://<yourusername>.github.io/my-book/jupyterlite/lab/index.html?path=files/how_to_use.ipynb)
[![Open in Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/yourusername/my-book/blob/main/content/files/how_to_use.ipynb)

## Step 1: Create Your Repository

- Go to this repository on GitHub.
- Click **"Use this template"** (green button).
- Name your repository (e.g., `my-book`) and create it.

## Step 2: Clone and Configure

- Clone your repository:
-
  ```bash
  git clone https://github.com/yourusername/my-book.git
  cd my-book
  ```

- Edit `_config.yml` to set your book’s details:
-
  ```yaml
  title: Your Book Title
  author: Your Name
  logo: path/to/your/logo.png  # Optional
  ```

- Edit `_toc.yml` to define your book’s structure:
-
  ```yaml
  format: jb-book
  root: intro
  chapters:
    - file: content/files/course_introduction
    - file: content/files/how_to_use
    - file: content/files/roadmap_prereqs
  ```

## Step 3: Generate and Customize Notebooks

- Run the script to create empty notebooks from `_toc.yml`:
  ```bash
  python notebooks/auto_notebook_creation_using_toc.py
  ```
- Check that notebooks appear in `content/files/`:
  ```bash
  ls content/files/*.ipynb
  ```
- Edit notebooks in `content/files/` (e.g., `intro.ipynb`) to add your content.

## Step 4: Deploy Your Book

- Set up GitHub Pages:
  - Go to Settings > Pages in your repository.
  - Set **Source** to **GitHub Actions**.
- Commit and push changes:
  ```bash
  git add _config.yml _toc.yml content/files/*.ipynb
  git commit -m "Set up my book"
  git push origin main

#### Your book will auto-deploy to `https://<yourusername>.github.io/my-book/jupyterlite/lab/`. Updates to `content/files/` trigger automatic rebuilds.
